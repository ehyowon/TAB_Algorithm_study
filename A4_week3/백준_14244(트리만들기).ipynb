# **트리 만들기**

## **문제**
n과 m이 주어졌을 때, n개의 노드로 이루어져 있고, m개의 리프로 이루어져 있는 트리를 만드는 프로그램을 작성하시오.

항상 정답이 존재하는 경우만 입력으로 주어진다.

트리는 사이클이 없는 연결 그래프이고, 리프는 차수가 1인 노드를 의미한다.

## **입력**
첫째 줄에 n과 m이 주어진다. (3 ≤ n ≤ 50, 2 ≤ m ≤ n-1)

## **출력**
첫째 줄부터 n-1개의 줄에 트리의 간선 정보를 출력한다. 트리의 정점은 0번부터 n-1번까지 이다.

## **코드**


리프노드가 루트노드를 포함한다는 것! 기억하기

여러 출력값이 나올 수 있는 문제이다 -> 오류 아닌가?

예를 들어 4 2 일 때 
> 0 = 1 = 2 = 3 -> 0,1 / 1,2 / 2,3

> 0 = 1 
>
> ㅤ= 2 = 3 -> 0,1 / 0,2 / 2,3

이렇게 두 가지가 나올 수 있다

하지만 이 문제에서는 하나의 입력이 주어졌을 때 여러 가지 트리를 구성할 수 있는 문제라서 출력값이 하나로 특정된 게 아니라 정답이 될 수 있는 여러 출력값 중 하나를 만족하면 맞다. 이게 정말 신기했던 점.

```python
n, m = map(int, input().split())

for i in range(n-1):
	if i < n-m:
		print(i, i+1)
	else:
		print(n-m, i+1)
```

내가 작성한 코드는 리프 전까지 맨 밑에 하나씩 붙이는 방법 !

즉, n-m 직전까지는 0-1, 1-2 , 2-3 이렇게 이어지다가 m개의 리프가 필요하다면 n-m 노드에 하나씩 노드들을 붙여서



이렇게 구현한다면, n개의 노드에 m개의 리프를 가진 트리를 만든 게 된다
